```python
# Sample Data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

# Create figure
plt.figure(figsize=(10, 6))

# First subplot (Top-left)
plt.subplot(2, 2, 1)
plt.plot(x, y1, color='r')
plt.title("Sine Wave")

# Second subplot (Top-right)
plt.subplot(2, 2, 2)
plt.plot(x, y2, color='b')
plt.title("Cosine Wave")

# Third subplot (Bottom-left)
plt.subplot(2, 2, 3)
plt.plot(x, y3, color='g')
plt.title("Tangent Wave")
plt.ylim(-5, 5)  # Limit y-axis to avoid large values

# Fourth subplot (Bottom-right)
plt.subplot(2, 2, 4)
plt.plot(x, y4, color='purple')
plt.title("Exponential Decay")

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Sample Data
x = np.linspace(0, 10, 100)
y1, y2, y3, y4 = np.sin(x), np.cos(x), np.tan(x), np.exp(-x)

# Create a figure and axes
fig, axes = plt.subplots(2, 2, figsize=(10, 6))

# First subplot
axes[0, 0].plot(x, y1, 'r')
axes[0, 0].set_title("Sine Wave")

# Second subplot
axes[0, 1].plot(x, y2, 'b')
axes[0, 1].set_title("Cosine Wave")

# Third subplot
axes[1, 0].plot(x, y3, 'g')
axes[1, 0].set_title("Tangent Wave")
axes[1, 0].set_ylim(-5, 5)  # Limit y-axis

# Fourth subplot
axes[1, 1].plot(x, y4, 'purple')
axes[1, 1].set_title("Exponential Decay")

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Sample data
t = np.linspace(0, 10, 100)
x = np.sin(t)
y = np.cos(t)
z = t

# Create 3D figure
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot 3D line
ax.plot3D(x, y, z, color='red', label="3D Line")
ax.set_title("Basic 3D Line Plot")

# Labels
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

# Show plot
plt.legend()
plt.show()
```

```python
# Generate random data
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

# Create 3D figure
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter3D(x, y, z, c=z, cmap='viridis', marker='o')

# Labels and title
ax.set_title("3D Scatter Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

# Show plot
plt.show()
```

```python
# Generate grid
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))  # Function for surface

# Create 3D figure
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
surf = ax.plot_surface(X, Y, Z, cmap='plasma')

# Color bar
fig.colorbar(surf)

# Labels
ax.set_title("3D Surface Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

# Show plot
plt.show()
```

```python
# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])
z = np.zeros(5)  # Start at zero
dx = dy = np.ones(5)  # Bar width
dz = [3, 5, 7, 6, 4]  # Bar heights

# Create 3D figure
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# 3D Bar plot
ax.bar3d(x, y, z, dx, dy, dz, color='skyblue')

# Labels
ax.set_title("3D Bar Chart")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Value")

# Show plot
plt.show()
```
