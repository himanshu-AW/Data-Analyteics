```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from mpl_toolkits.mplot3d import Axes3D  # For 3D plots

data = {
    'StudyTime': [2, 3, 1, 4, 2.5, 3.5, 1.5, 4.5, 3, 2.2],
    'SleepHours': [7, 8, 6, 9, 7.5, 8.5, 6.5, 9.5, 8, 7],
    'ExamScore': [60, 75, 55, 90, 70, 85, 50, 95, 80, 65]
}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8, 6)) # Plot size
plt.scatter(df['StudyTime'], df['ExamScore'], c=df['SleepHours'], cmap='viridis', s=50) # Color by SleepHours, size adjusted
plt.xlabel('Study Time (hours)')
plt.ylabel('Exam Score')
plt.title('Study Time vs. Exam Score (Colored by Sleep Hours)')
plt.colorbar(label='Sleep Hours') # Add a colorbar
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
plt.scatter(df['StudyTime'], df['ExamScore'])
plt.title('Study Time vs. Exam Score')

plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
plt.scatter(df['SleepHours'], df['ExamScore'])
plt.title('Sleep Hours vs. Exam Score')

#plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['StudyTime'], df['SleepHours'], df['ExamScore'], c=df['ExamScore'], cmap='coolwarm')
ax.set_xlabel('Study Time')
ax.set_ylabel('Sleep Hours')
ax.set_zlabel('Exam Score')
plt.title('3D Plot of Study Time, Sleep Hours, and Exam Score')
plt.show()
```
